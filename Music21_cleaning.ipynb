```python
def load_data():
  main_dir = '/content/gdrive/MyDrive/Composers'
  notes = []
  rhythms = []
  for composer in os.listdir(main_dir)[1:3]: # we need [1:] in order to not get the .DS_store
    print(composer)
    print(os.listdir(main_dir + '/' + composer))
    for song in os.listdir(main_dir + '/' + composer)[1:3]:
      midi = converter.parse(main_dir + '/' + composer + '/' + song)
      parse_notes = None
      parts = instrument.partitionByInstrument(midi)

      #return parts

      if parts: # file has instrument parts
        notes_to_parse = parts.parts[0].recurse()
      else: # file has notes in a flat structure
        notes_to_parse = midi.flat.notes

      

      # # places to store notes and rhythms
      # notes = []
      # rhythms = []

      for element in notes_to_parse:
        if isinstance(element,note.Note):
          print('entered')
          notes.append(str(element.pitch))
          rhythms.append(element.duration.quarterLength)
        elif isinstance(element,chord.Chord):
          notes.append('.'.join(str(n) for n in element.normalOrder))
          rhythms.append(element.duration.quarterLength)

  df = pd.DataFrame({"Note":notes,"Duration":rhythms})
  return df, notes, rhythms
```
